import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Carregar o arquivo Excel
data = pd.read_excel('seu_arquivo.xlsx')  # Substitua 'seu_arquivo.xlsx' pelo nome do seu arquivo

# Selecionar todas as colunas necessárias para a clusterização
data_values = data.values

# Aplicar o Min-Max Scaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data_values)

# Aplicar o algoritmo de clusterização (DBSCAN)
dbscan = DBSCAN(eps=0.3, min_samples=2)
labels = dbscan.fit_predict(data_scaled)

# Reduzir a dimensionalidade dos dados usando PCA para visualização
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Visualizar os resultados
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=labels)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Clusterização com DBSCAN e PCA')
plt.colorbar(label='Cluster')
plt.show()
